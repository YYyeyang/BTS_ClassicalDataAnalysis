```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
df=pd.read_csv("Y:/GitHub/BTS_MasterInBigData-Francisco/Session_13_DSF/1_titanic_dataset.csv")
df.head()
```

```python
new_age=np.where(df['Age'].isnull(),28,df['Age'])
df['Age']=new_age
```

```python
df.isnull().sum()
```

```python
label_encoder = preprocessing.LabelEncoder()
encoded_sex = label_encoder.fit_transform(df["Sex"])
df['Sex'] = encoded_sex
```

```python
encoded_Ticket = label_encoder.fit_transform(df["Ticket"])
df['Ticket'] = encoded_Ticket
```

```python
df.info()
```

```python
df.drop(['PassengerId'], axis=1).corr(method='pearson').style.format("{:.1}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```
